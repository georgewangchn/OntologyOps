## 2. 构建本体

定义核心类（疾病、症状、物种...），从 CSV 加载疾病层次结构。

````python
onto = create_ontology()
onto = load_diseases(onto)
print(f'✅ 已加载 {len(list(onto.classes()))} 个类')
````

## 3. 加载症状关系

## 4. 执行推理

调用 HermiT 推理机，输入一个病例，输出疑似疾病列表。

````python
from reasoner import diagnose, load_ontology, print_diagnosis
import json

case = {
    'pet_type': 'cat',
    'symptoms': ['发热', '呕吐', '腹泻'],
    'breed': '英短',
    'age': 2
}

onto = load_ontology()
results = diagnose(onto, case)
print_diagnosis(results)
````
